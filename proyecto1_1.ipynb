```python
weather_files = sorted(glob('weather*.txt'))
weather_files #invoca los nombres de los archivos ordenados con sorted, éste es solo para visualizar el comportamiento

#separar las columnas de cada archivo

# Leer el archivo .txt en un DataFrame
df = pd.read_csv('weather01.txt', header=None)  # 'header=None' ya que no hay encabezado

# Separar los datos en 7 columnas usando el delimitador adecuado
df_separado = df[0].str.split(r'\s+', expand=True)  # Aquí '\s+' representa uno o más espacios

# Asignar los nombres de las columnas
df_separado.columns = ['indice','X', 'year', 'month', 'measure', 'X1', 'X2']

# Guardar el nuevo DataFrame en un archivo de salida
df_separado.to_csv('weather01_separado.txt', index=False, sep='\t')

# Imprimir el DataFrame resultante
print(df_separado)

#eliminar primera columna de cada archivo
# Leer el archivo .txt en un DataFrame
df = pd.read_csv('weather01_separado.txt', delimiter='\t')  # Usa el delimitador adecuado (tabulador o espacio)

# Eliminar la primera columna
df = df.iloc[:, 2:]

# Eliminar la primera fila ya que es solo para los nombres de las columnas
df = df.drop(index=0)

# Guardar el DataFrame sin la primera columna en un nuevo archivo
df.to_csv('newweather01.txt', index=False)

pd.read_csv('newweather01.txt')
```

```python
newdf_melt = newdf_concatenado.melt(id_vars=['year','month','measure'])
newdf_melt
```

```python
newdf_melt.dropna(inplace=True) #elimina filas con valores nulos
newdf_melt
```

```python
newdf_melt.rename(columns ={'variable':'day'},inplace=True)
newdf_melt['day'] = newdf_melt['day'].str.extract('(\d+)').astype(int) #cambia el tipo de dato de "day"
newdf_melt['date'] = pd.to_datetime(newdf_melt[['year','month','day']])
newdf_melt
```

```python
newdf_melt.drop(['year','month','day'],axis=1,inplace=True) #eliminar las columnas innecesarias
newdf_melt
```

```python
df1 = newdf_melt.pivot(index='date',columns='measure',values='value').reset_index()
df1
```
